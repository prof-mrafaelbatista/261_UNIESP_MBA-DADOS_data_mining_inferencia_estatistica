## 📌 1. Basic Data Types

In Python, every value has a **type**. Understanding types is fundamental because they determine which operations are valid and how data is stored in memory.

| Type | Keyword | Example | Common Use in Data Science |
|------|---------|---------|---------------------------|
| Text | `str` | `"Data Mining"` | Column names, categories, labels |
| Integer | `int` | `2024` | Counts, IDs, year values |
| Decimal | `float` | `3.14159` | Metrics, probabilities, measurements |
| Boolean | `bool` | `True` / `False` | Flags, filters, binary classification |

> 💡 **Curiosity:** Python is *dynamically typed* — you never declare the type of a variable. The interpreter infers it automatically at runtime. This is different from Java or C, where you must write `int x = 5;`.

````python
disciplina = "Data Mining" # String
````

````python
ano = 2024 # Integer
````

````python
pi = 3.14159 # Float
````

````python
cadastro_ativo = True # Boolean 
````

````python
print(type(disciplina), type(ano), type(pi), type(cadastro_ativo))
````

````python
# ── Type conversion (casting) ──────────────────────────────────────────────────
# Very common when loading data from CSVs — values often arrive as strings

idade_texto = "25"                  # arrives as a string from a CSV file
idade_numero = int(idade_texto)     # convert to integer for arithmetic

print(f"Before conversion: {idade_texto!r}  → type: {type(idade_texto).__name__}")
print(f"After  conversion: {idade_numero!r} → type: {type(idade_numero).__name__}")

# ── None — represents the absence of a value (like NULL in SQL / NaN concept) ──
valor_ausente = None
print(f"\nNone type : {type(valor_ausente)}")
print(f"Is None?  : {valor_ausente is None}")   # always use `is`, not `==`, for None

# ── f-strings — the modern and recommended way to format strings ────────────────
nome  = "Maria"
curso = "Data Science"
print(f"\nStudent {nome} is enrolled in {curso}.")
````

## 📋 2. Lists

A **list** is an ordered, mutable collection of items. It is one of the most frequently used data structures in Python and maps directly to a *column of values* in a dataset.

```python
# Syntax
my_list = [element1, element2, element3]
```

**Key characteristics:**
- ✅ **Ordered** — elements maintain their insertion order
- ✅ **Mutable** — you can add, remove, or modify elements after creation
- ✅ **Allows duplicates**
- ✅ **Zero-indexed** — the first element is at index `0`

**Indexing cheat-sheet:**

| Expression | Meaning |
|------------|---------|
| `list[0]` | First element |
| `list[-1]` | Last element |
| `list[1:3]` | Elements at index 1 and 2 (end is exclusive) |
| `list[:3]` | First 3 elements |
| `list[2:]` | From index 2 to the end |

> 💡 **Curiosity:** Python lists are internally implemented as *dynamic arrays*. Accessing an element by index (`list[i]`) is **O(1)** — instant, regardless of the list size! However, inserting at the beginning is O(n) because all elements must shift.

````python
notas = [8.5, 9.0, 7.5, 10.0] # List of Floats
````

````python
print(notas[0]) # Acessando o primeiro elemento da lista
````

````python
print(notas[-1]) # Acessando o último elemento da lista
````

````python
print(notas[1:3]) # Acessando uma fatia da lista (elementos 1 e 2)
````

````python
notas.append(9.5) # Adicionando um novo elemento à lista
````

````python
print(len(notas)) # Imprimindo o número de elementos na lista
````

````python
# ── List comprehension ─────────────────────────────────────────────────────────
# A concise, Pythonic way to build new lists — heavily used in preprocessing

# Normalize grades to a 0–1 scale
notas_normalizadas = [nota / 10.0 for nota in notas]
print("Normalized grades:", notas_normalizadas)

# Filter: keep only passing grades (>= 7.0)
aprovados = [nota for nota in notas if nota >= 7.0]
print("Passing grades   :", aprovados)

# ── Useful list methods ────────────────────────────────────────────────────────
notas_copia = notas.copy()          # always copy before sorting to preserve original

notas_copia.sort()                  # sort ascending (in-place)
print("\nSorted (asc) :", notas_copia)

notas_copia.sort(reverse=True)      # sort descending (in-place)
print("Sorted (desc):", notas_copia)

print(f"\nSum   : {sum(notas)}")
print(f"Mean  : {sum(notas) / len(notas):.2f}")  # :.2f formats to 2 decimal places
````

## 🗂️ 3. Dictionaries

A **dictionary** stores data as **key–value pairs**. Think of it as a two-column table: a unique identifier (key) and the associated data (value). In data science, a single row of a dataset can be represented as a dictionary.

```python
# Syntax
my_dict = {"key1": value1, "key2": value2}
```

**Key characteristics:**
- ✅ **Keys must be unique** — duplicate keys overwrite previous values
- ✅ **Values can be any type** — including lists, other dicts, or even functions
- ✅ **Fast lookups** — accessing a value by key is O(1) on average (uses a hash table internally)
- ✅ **Insertion-ordered** — Python 3.7+ guarantees keys appear in insertion order

> 💡 **Curiosity:** `pandas` DataFrames are deeply connected to dictionaries. `pd.DataFrame([dict1, dict2, ...])` builds a full table from a list of dictionaries — where each dict is a row and the keys become column names.

````python
aluno = {
    "nome": "Maria",
    "idade": 22,
    "curso": "Ciência de Dados",
    "notas": [8.5, 9.0, 7.5, 10.0],
    "aprovado": True
}
````

````python
print(aluno['nome']) # Acessando o nome do aluno
````

````python
print(aluno.keys()) # Imprimindo as chaves do dicionário
````

````python
for chave, valor in aluno.items():
    print(f"{chave}: {valor}") # Imprimindo cada chave e valor do dicionário
````

````python
# ── Safe access with .get() ────────────────────────────────────────────────────
# Avoids KeyError when the key might not exist — always prefer .get() in pipelines

email = aluno.get("email", "not provided")   # returns default value if key is absent
print(f"Email: {email}")

# ── Adding and updating entries ────────────────────────────────────────────────
aluno["email"] = "maria@uniesp.edu.br"       # add a new key
aluno["notas"].append(8.0)                   # update a nested list in place
print(f"\nUpdated record: {aluno}")

# ── Building a dict from two lists with zip() ──────────────────────────────────
# Useful when constructing records from separate column arrays

campos  = ["nome",    "nota_final", "aprovado"]
valores = ["Carlos",  8.7,           True]

registro = dict(zip(campos, valores))        # pairs each field name with its value
print(f"\nNew record from zip: {registro}")
````

## 🔁 4. Control Flow

Control flow structures define the **logic and decisions** in your code. In data science they appear everywhere — filtering rows, applying conditional transformations, and iterating over datasets.

### `for` loop
Iterates over any *iterable*: lists, dict keys/values, ranges, DataFrame rows...

### `if / elif / else`
Applies conditional logic based on data values — essential for feature engineering and data validation.

> 💡 **Curiosity:** Python's `for` is actually a **for-each** loop (similar to `foreach` in Java or C#). It always iterates over the *items* of an iterable, not over indices. When you need both the index and the value, use `enumerate()`.

````python
for nota in notas:
    if nota >= 7.0:
        print(f"Nota {nota} é considerada boa.")
    else:
        print(f"Nota {nota} é considerada ruim.")
````

````python
# ── enumerate() — iterate with index and value simultaneously ──────────────────
print("=== Grades with index ===")
for i, nota in enumerate(notas):
    print(f"  [{i}] {nota}")

# ── Multi-branch classification with elif ─────────────────────────────────────
# Simulates a grading scale — a common feature engineering pattern

print("\n=== Grade classification ===")
for nota in notas:
    if nota >= 9.0:
        conceito = "A — Excellent"
    elif nota >= 7.0:
        conceito = "B — Good"
    elif nota >= 5.0:
        conceito = "C — Pass"
    else:
        conceito = "F — Fail"
    print(f"  {nota:.1f} → {conceito}")
````

## ⚙️ 5. Functions

A **function** is a reusable, named block of code that performs a specific task. In data science, functions are the building blocks of **data pipelines** — each step (cleaning, transformation, analysis) is typically encapsulated in a function.

```python
# Syntax
def function_name(parameter1, parameter2):
    """Docstring — describes what the function does."""
    # body
    return result
```

**Why write functions?**
- 🔁 **Reusability** — write once, call many times with different inputs
- 🧪 **Testability** — easier to isolate and validate individual logic pieces
- 📖 **Readability** — a well-named function is self-documenting
- 🏗️ **Pipelines** — chain functions to build complete data workflows

> 💡 **Curiosity:** In Python, functions are **first-class objects** — they can be assigned to variables, passed as arguments to other functions, and returned as values. This is exactly what makes `df.apply(my_function)` in pandas work!

````python
def resumo_lista(valores):
    """
    Compute a basic statistical summary of a numeric list.

    Parameters
    ----------
    valores : list of float
        Numeric values to summarize. Must not be empty.

    Returns
    -------
    dict
        Dictionary with three keys:
        - 'media'  : arithmetic mean of the values
        - 'minimo' : smallest value in the list
        - 'maximo' : largest value in the list
    """
    return {
        "media": sum(valores) / len(valores),
        "minimo": min(valores),
        "maximo": max(valores)
    }
````

````python
print(resumo_lista(notas)) # Imprimindo o resumo da lista de notas
````

````python
def classificar_nota(nota, limiar=7.0):
    """
    Classify a single grade as passing or failing.

    Parameters
    ----------
    nota : float
        Grade value to classify (expected range: 0.0 to 10.0).
    limiar : float, optional
        Passing threshold. Defaults to 7.0.

    Returns
    -------
    str
        'Pass' if nota >= limiar, 'Fail' otherwise.
    """
    return "Pass" if nota >= limiar else "Fail"


def estatisticas_completas(valores):
    """
    Compute extended descriptive statistics for a list of numeric values.

    Parameters
    ----------
    valores : list of float
        Non-empty list of numeric values to analyse.

    Returns
    -------
    dict
        Dictionary with keys: count, mean, min, max, range.
    """
    n      = len(valores)
    media  = sum(valores) / n
    minimo = min(valores)
    maximo = max(valores)
    return {
        "count": n,
        "mean" : round(media, 2),
        "min"  : minimo,
        "max"  : maximo,
        "range": round(maximo - minimo, 2),   # spread of the distribution
    }


# ── Applying the functions ─────────────────────────────────────────────────────
print("=== Grade classification ===")
for nota in notas:
    print(f"  {nota} → {classificar_nota(nota)}")

print("\n=== Extended statistics ===")
stats = estatisticas_completas(notas)
for chave, valor in stats.items():
    print(f"  {chave:<8}: {valor}")
````
